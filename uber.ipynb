```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import root_mean_squared_error
```

```python
#data taken from around 1pm saturday
df = pd.read_csv("Book1.csv")

X = df[['distance', 'time']]
y = df['price']

model = LinearRegression()
model.fit(X, y)

distanceP = model.coef_[0]
timeP = model.coef_[1]
intercept = model.intercept_

print("distanceP (per mile):", distanceP)
print("timeP (per minute):", timeP)
print("intercept (baseFare + fees):", intercept)
```

```python
y_pred = model.predict(X)
rmse = root_mean_squared_error(y, y_pred)

print("RMSE:", rmse)
```
